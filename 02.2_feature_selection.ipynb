```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./Source/feature_all.csv', index_col='pair_id')
df.drop(columns='mra', inplace=True)
df.head()
```

```python
from sklearn.metrics import precision_recall_curve, confusion_matrix
```

```python
best_features = []
best_score = []
b = 1
flag = True

max_fscore = 0

while flag:
    check_columns = set(df.columns[2:-1]) - set(best_features)
    print(check_columns)
    
    flag = False
    
    best_feature = 0
    
    fscore = 0
    mcc_value = 0
    for col in check_columns:
        temp_df = df[best_features + [col]].copy()
        temp_df['mean'] = temp_df.mean(axis=1)

        precision, recall, thresholds = precision_recall_curve(df.is_duplicate, temp_df['mean'])
        fscore = (1 + b**2) * precision * recall / (b**2 * precision + recall)
        ix = np.argmax(fscore)

#         print(col, fscore[ix])

        if fscore[ix] > max_fscore:
            max_fscore = fscore[ix]
            best_feature = col
            flag = True

    if flag:
        best_features.append(best_feature)
        best_score.append(max_fscore)
    
    print()
    print(best_feature, max_fscore)
    print(best_features)
```

```python
precision, recall, thresholds = precision_recall_curve(df.is_duplicate, df.mean)
b = 0.5 # 0.7
fscore = (1 + b**2) * precision * recall / (b**2 * precision + recall)
ix = np.argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f, Precision=%.3f, Recall=%.3f' % (thresholds[ix], 
                                                                        fscore[ix],
                                                                        precision[ix],
                                                                        recall[ix]))
```
